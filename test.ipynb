```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csvdir = 'csvfiles/11_data921_rates10.csv'
df = pd.read_csv(csvdir).mean()

num_rates = 10
bpp1 = np.array(
    [df['R'+str(idx)+'_bpp'] for idx in range(num_rates)]).round(4)
psnr1 = np.array(
    [df['R'+str(idx)+'_mseF,PSNR (p2point)'] for idx in range(num_rates)]).round(4)
print('11', '\n', bpp1, '\n', psnr1)
```

```python
csvdir = 'csvfiles/12_data1061_rates10.csv'
df = pd.read_csv(csvdir).mean()

num_rates = 10
bpp2 = np.array(
    [df['R'+str(idx)+'_bpp'] for idx in range(num_rates)]).round(4)
psnr2 = np.array(
    [df['R'+str(idx)+'_mseF,PSNR (p2point)'] for idx in range(num_rates)]).round(4)
print('11', '\n', bpp2, '\n', psnr2)
```

```python
fig, ax = plt.subplots(figsize=(7.3*2,5*2))

plt.plot(bpp1, psnr1, label='11', color='red',marker='^', 
         linewidth=3,markersize=12)
plt.plot(bpp2, psnr2, label='12', color='blue', marker='o', 
         linewidth=3,markersize=12)

ax.locator_params(axis='x', nbins=20)
ax.locator_params(axis='y', nbins=20)
plt.tick_params(labelsize=36)
plt.xlabel("bpp", fontdict={'family':'serif', 'weight':'normal','size':40})
plt.ylabel("D1 PSNR (dB)", fontdict={'family':'serif', 'weight':'normal','size':40})

plt.title('KITTI', fontdict={'family':'serif', 'weight':'normal','size':60})
plt.grid(ls='-.')
plt.legend(loc='lower right', ncol=1,
           prop={'family':'serif', 'weight':'normal','size':32})
fig.tight_layout()
os.makedirs('figs', exist_ok=True)
fig.savefig(os.path.join('figs', 'kitti.png'))
```
