```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csvdir = 'csvfiles/q1mm_data1_rates10.csv'
df = pd.read_csv(csvdir).mean()
num_rates = 10
bpp1 = np.array([df['R'+str(idx)+'_bpp'] for idx in range(num_rates)]).round(4)
psnr1 = np.array([df['R'+str(idx)+'_mseF,PSNR (p2point)'] for idx in range(num_rates)]).round(4)
print(bpp1, '\n', psnr1)
```

```python
csvdir = 'csvfiles/q16bit_data1_rates10.csv'
df = pd.read_csv(csvdir).mean()
num_rates = 10
bpp2 = np.array([df['R'+str(idx)+'_bpp'] for idx in range(num_rates)]).round(4)
psnr2 = np.array([df['R'+str(idx)+'_mseF,PSNR (p2point)'] for idx in range(num_rates)]).round(4)
print(bpp2, '\n', psnr2)
```

```python
fig, ax = plt.subplots(figsize=(7.3*2,5*2))
plt.plot(bpp1, psnr1, label='q16bit', color='red',marker='^', linewidth=4,markersize=16)
plt.plot(bpp2, psnr2, label='q1mm', color='blue', marker='v', linewidth=4,markersize=16)
ax.locator_params(axis='x', nbins=10)
ax.locator_params(axis='y', nbins=10)
plt.tick_params(labelsize=36)
plt.xlabel("bpp", fontdict={'size':40})
plt.ylabel("PSNR", fontdict={'size':40})
plt.title('kitti_seqs11_000000.bin', fontdict={'size':60})
plt.grid(ls='-.')
plt.legend(loc='lower right', prop={'size':32})
fig.tight_layout()
os.makedirs('figs', exist_ok=True)
fig.savefig(os.path.join('figs', 'rdcurve.png'))
```
